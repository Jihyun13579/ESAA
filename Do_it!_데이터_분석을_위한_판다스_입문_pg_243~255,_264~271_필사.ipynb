```python
# Date>datetime 오브젝트로 변환하여 ebola 데이터 프레임 다시 생성함

ebola = pd.read_csv('/content/country_timeseries.csv', parse_dates=['Date'])
print(ebola.head().iloc[:,:5])
```

```python
#Date을 인덱스로 지정 후 최댓값-최솟값으로 시간범위 지정하여 new_idx에 저장
#결과: 날짜가 아예 없었던 데이터의 인덱스 생성 가능

ebola.index = ebola['Date']
new_idx = pd.date_range(ebola.index.min( ), ebola.index.max( ))
```

```python
#new_idx 살펴보기 > ebola 시간순서와 반대로 되어 있음 >reversed 메서드 사용

print(new_idx)
```

```python
new_idx = reversed(new_idx)
```

```python
#new_idx를 새로운 인덱스로 지정 : reindex 이용

ebola = ebola.reindex(new_idx) 
print(ebola.head( ).iloc[:, :5])
```

```python
print(ebola.tail( ).iloc[:, :5])
```

```python
# 각 나라의 에볼라 발병일 옮기기

last_valid = ebola.apply(pd.Series.last_valid_index) 
#last_valid_index: 가장 오래된 데이터 가져옴
print(last_valid)
```

```python
first_valid = ebola.apply(pd.Series.first_valid_index) 
#first_valid_index: 가장 최근의 데이터 가져옴
print(first_valid)
```

```python
# (가장 처음 발병한 날 - 각 나라의 에볼라 발병일)한 만큼 각 나라의 에볼라 발병일 옮기기
earliest_date = ebola.index.min( ) 
print(earliest_date)
```

```python
shift_values = last_valid - earliest_date
print(shift_values)
```

```python
# 각 나라의 에볼라 발병일 옮기기
# shift 메서드(:인잣값만큼 데이터 밀어내는 메서드) 사용 > 모든 열의 값을 shift_values 만큼 옮기기

ebola_dict = {}
for idx, col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d) 
    ebola_dict[col] = shifted
```

```python
# ebola_dict : 시간 다시 설정한 데이터가 딕셔너리 형태로 저장되어 있음 > 데이터프레임으로 변환

ebola_shift = pd.DataFrame(ebola_dict)
```

```python
# 에볼라 최초 발병일 기준으로 모든 열의 데이터 옮겨졌는지 확인

print(ebola_shift.tail())
```

```python
# 인덱스 : Day 열로 지정 후 필요없는 Date, Day열 삭제 > 그래프 그릴 dataframe 완성

ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Date', 'Day'], axis=1)

print(ebola_shift.tail())
```

```python
# 그래프 그리기

fig, ax = plt.subplots()
ax = ebola_shift.iloc[:,:].plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()
```
