### 텍스트 토큰화
- 문장 토큰화 : 문서에서 문장 분리
- 단어 토큰화 : 문장에서 단어 토큰으로 분리

#### 문장 토큰화(sentence tokenization)
- 문장의 마침표(,), 개행문자(\n) 등 문장의 마지막을 뜻하는 기호에 따라 분리하는 것이 일반적
- 정규 표현식에 따른 문장 토큰화도 가능

- NLTK의 sent_tokenize 이용해 문장 토큰화 진행하기
 - 3개의 문장으로 이루어진 텍스트 문서 > 문장으로 각각 분리하는 예제
 - NLTK : 단어 사전 등 참조 필요 데이터셋 인터넷으로 다운로드 가능
 - sent_tokenize() : 각각의 문장으로 구성된 list 객체 반환
 - 각 문장이 가지는 시맨틱한 의미가 중요한 요소로 사용될 때 사용

````python
from nltk import sent_tokenize
import nltk
nltk.download('punkt')   # 마침표, 개행 문자 등의 데이터 세트 다운로드

text_sample = 'The Matrix is everywhere its all around us, here even in this room. \
You can see it out your window or on your television. \
You feel it when you go to work, or go to churxh or pay your taxes. '
sentences = sent_tokenize(text=text_sample)
````

````python
# 반환된 list 객체 : 3개의 문장으로 된 문자열 보유
print(type(sentences), len(sentences))
print(sentences)
````

#### 단어 토큰화
: 문장을 단어로 토큰화하는 것
- 기본적으로 공백, 콤마(,), 마침표(.), 개행 문자 등으로 단어 분리 or 정규 표현식으로 다양한 유형으로 토큰화 수행
- 문장 분리하는 구분자(마침표, 개행문자)로 단어 토큰화 가능 > 단어의 순서가 중요하지 않은 경우, 문장 토큰화 상요하지 않고 단어 토큰화만 사용해도 충분

- NLTK의 word_tokennize 이용해 단어 토큰화 진행하기

````python
from nltk import word_tokenize

sentence = "The Matrix is everywhere its all around us, here even in this room."
words = word_tokenize(sentence)
print(type(words), len(words))
print(words)
````

- sent_tokennizedhk word_tokennize를 조합해 문서에 대해서 모든 단어 토큰화해보기
 - 문장 토큰화된 text_sampel(3개의 문장으로 구성)에 문장별로 단어 토큰화 적용

````python
from nltk import word_tokenize, sent_tokenize

# 여러 개의 문장으로 된 입력 데이터를 문장별로 단어 토큰화하게 만드는 함수 생성
def tokenize_text(text):

    # 문장별로 분리 토큰
    sentences = sent_tokenize(text)
    # 분리된 문장별 단어 토큰화
    word_tokens = [word_tokenize(sentence) for sentence in sentences]
    return word_tokens

# 여러 문장에 대해 문장별 단어 토큰화 수행.
word_tokens = tokenize_text(text_sample)
print(type(word_tokens), len(word_tokens))
print(word_tokens)
 # word_tokens : 3개의 리스트 객체 내포하는 리스트, 각 리스트 : 문장별로 토큰화된 단어 요소로 가지고 있음
````

- 문장을 단어별로 하나씩 토큰화 시, 문맥적인 의미 무시될 수밖에 없음

**n_gram** : 연속된 n개의 단어를 하나의 토큰화 단위로 분리해내는 것
- n개 단어 크기 윈도우 만들어 문장의 처음부터 오른쪽으로 움직이면서 토큰화 수행
- ex. "Agent Smith knocks the door" : 2-gram(=bigram) > (Agent, Smith), (Smith, knocks), (knocks, the), (the, door) 연속된 2개의 단어들 순차적으로 이동하면서 단어들 토큰화해줌


### 스톱 워드 제거

스톱 워드(Stop word) : 분석에 의미가 없는 단어 지칭
 - ex. is, the, a, will... > 필수 문법 요소, but 문맥적으로 큰 의미 X
 - 빈번하게 나타나는 스톱 워드들 > 중요한 단어로 인지될 수 있으므로 사전 삭제 필요 (중요한 전처리 작업)
 - 언어별로 스톱 워드 목록화되어 있음
  - NLTK : 가장 다양한 언어의 스톱 워드 제공

- NTTK의 스톱워드에 어떤 것이 있을지 확인하기
 - NLTK의 stopwords 목록 내려받기

````python
import nltk
nltk.download('stopwords')
````

 - 몇 개의 stopwords 있는지 알아보기 + 처음 20개 확인하기

````python
print('영어 stop words 개수:', len(nltk.corpus.stopwords.words('english')))
print(nltk.corpus.stopwords.words('english')[:20])
# 영어 : 스톱워드 개수 179개
````

- 위의 문장 토큰화, 단어 토큰화한 word_tokens 리스트에 대하여 stopwords 필터링으로 제거 > 분석을 위한 의미 있는 단어만 추출하기

````python
import nltk

stopwords = nltk.corpus.stopwords.words('english')
all_tokens = []
# 위 예제에서 3개의 문장별로 얻은 word_tokens list에 대해 스톱 워드를 제거하는 반복문
for sentence in word_tokens:
    filtered_words=[]
    # 개별 문장별로 토큰화된 문장 list에 대해 스톱 워드를 제거하는 반복문
    for word in sentence:
        # 소문자로 모두 변환합니다.
        word = word.lower()
        # 토큰화된 개별 단어가 스톱 워드의 단어에 포함되지 않으면 word_tokens에 추가
        if word not in stopwords:
            filtered_words.append(word)
    all_tokens.append(filtered_words)

print(all_tokens)
# is 등 스톱워드 제거되었음을 확인 가능
````

### Stemming과 Lemmatization
: 문법적, 의미적으로 변화하는 단어의 원형을 찾는 것
 - 시제, 진행형, 완료형, 인칭, 단수/복수에 따라 다양하게 변화하는 단어의 원형 단어 찾기
 - 두 기능 모두 원형 단어 찾는다는 목적
 - Lemmatization : Stemming보가 정교, 의미론적 기반에서 단어의 원형 찾음, 품사와 같은 문법적인 요소와 더 의미적인 부분 감안해 정확한 철자로 된 어근 단어 찾아줌
 - Stemming: 원형 단어로 변환 시 일반적인 방법을 적용 or 더 단순화된 방법 적용해 원래 단어에서 일부 철자가 훼손된 어근 단어 추출하는 경향
 - NLTK : 다양한 Stemmer 제공 / Lemmatization 위해 WordNetLemmatizer 제공


- NLTK 이용하여 Stemming과 Lemmatization 비교하기
 - NLTK의 LancasterStemmer 이용해 Stemmer 살펴보기

````python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
# LancasterStemmer() stemmer 객체 생성 후, 이 객체의 stem('단어') 메서드 호출
# 괄호 안에 들어간 '단어'의 Stemming
print(stemmer.stem('working'), stemmer.stem('works'), stemmer.stem('worked'))  # 제대로 인식 O
print(stemmer.stem('amusing'), stemmer.stem('amuses'), stemmer.stem('amused')) # 제대로 인식 X

print(stemmer.stem('happier'), stemmer.stem('happiest'))  # 제대로 인식 X
print(stemmer.stem('fancier'), stemmer.stem('fanciest'))  # 제대로 인식 X
````

- NLTK의 WordNetLemmatizer 이용해 Lemmatization 수행하기
 - 단어의 '품사' 입력 필요 : 동사-'v', 형용사-'a'

````python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

lemma = WordNetLemmatizer()
print(lemma.lemmatize('amusing', 'v'), lemma.lemmatize('amuses', 'v'), lemma.lemmatize('amused','v')) # 제대로 인식 X
print(lemma.lemmatize('happier', 'a'), lemma.lemmatize('happiest', 'a')) # 제대로 인식 X
print(lemma.lemmatize('fancier', 'a'), lemma.lemmatize('fanciest', 'a')) # 제대로 인식 X
````

### BOW 피처 벡터화
**피처 벡터화** : 머신러닝 알고리즘 : 숫자형 피처를 데이터로 입력받아 동작 > 텍스트를 특정 의미를 가지는 숫자형 값인 벡터 값으로 변환하는 것
- 각 문서(Document)의 텍스트를 단어로 추출해 피처로 할당하고, 각 단어의 발생 빈도와 같은 값을 피처에 값으로 부여 > 각 문서를 이 단어 피처의 발생 빈고 값으로 구성된 벡터로 만드는 기법
- 기존 텍스트 데이터를 또 다른 형태의 조합으로 변경 > 넓은 범위의 피처 추출에 포함됨

BOW 모델의 피처 벡터화 : 모든 문서에서 모든 단어를 칼럼 형태로 나열하고 각 문서에서 해당 단어의 횟수나 정규되된 빈도를 값으로 부여하는 데이터 세트 모델로 변경하는 것

- BOW 피처 벡터와의 2가지 방식

1. 카운트 기반의 벡터화
 - 카운트 벡터화: 단어 피처에 값 부여 시 각 문서에서 해당 단어가 나타나는 횟수, Count 부여하는 경우
 - 카운트 값이 높음 > 중요한 단어
 - 언어의 특성상 문장에서 자주 상요되는 단어까지 높은 값 부여받는다는 단점

2. TF-IDF(Term Frequency - Inverse Document Frequency) 기반의 벡터화
 - 카운트 기반 벡터화의 단점 보완
 - 개별 문서에서 자주 나나타는 단어에 높은 가중치를 주되, 모든 문서에서 전반적으로 자주 나타나는 단어에 대해서 페널티 부여하는 방식으로 값 부여 < 모든 문서에 자주 나타나는 단어라면 해당 단어는 언어 특성상 범용적으로 자주 사용되는 단어일 가능성이 높기 때문
 - 문서마다 텍스트가 길고, 문서의 개수가 많은 경우, TF-IDF 방식이 더 좋은 예측 성능

### BOW 벡터화를 위한 희소 행렬
사이킷런의 CountVectorizer/TfidVectoerizer : 텍스트를 피처 단위로 벡터화해 변환, CSR 형태의 희소 행렬 반환


**희소 행렬**
- 모든 문서의 단어 추출해 피처 벡터화하는 방법 : 필연적으로 많은 변수(수만개~수십만개)의 피처 칼럼 생성, but, 대규모 행렬 생성되더라도 각 문서가 가지는 단어의 수 제한적 > 대규모 행렬의 대부분의 값을 0이 차지하는 행렬=희소행렬 생성됨
- BOW 형태 가진 언어 모델의 피처 벡터화 : 대부분 희소 행렬
- 불필요한 0값 메모리 공간에 할당 > 메모리 공간 많이 필요
- 행렬의 크기 금 > 연산 시에도 데이터 액세스 위한 시간 많이 소모됨
- 희소 행렬을 물리적으로 적인 메모리 공간 차지할 수 있도록 변환 필요 > **COO형식**과 **CSR형식**
 - CSR형식: 일반적으로 큰 희속 행렬 저장하고 계산 수행하는 능력 뛰어남 > 더 많이 사용

### 희소 행렬 - COO(Coordinate:좌표) 형식
: 0이 아닌 데이터만 별도의 데이터 배열(Array)에 저장하고, 그 데이터가 가리키는 행과 열의 위치를 별도의 배열로 저장하는 방식
- 주로 사이파이(Scipy) 이용


````python
import numpy as np
# 주어진 데이터
dense = np.array( [[3, 0, 1], [0, 2, 0]] )
````

````python
from scipy import sparse

# 0이 아닌 데이터 추출
data = np.array([3, 1, 2])

# 행 위치와 열 위치를 각각 배열로 생성
row_pos = np.array([0, 0, 1])
col_pos = np.array([0, 2, 1])

# sparse 패키지의 coo_matrix를 이용해 COO 형식으로 희소 행렬 생성
sparse_coo = sparse.coo_matrix((data, (row_pos, col_pos)))
````

````python
sparse_coo.toarray()
````

### 희소 행렬 - CSR 형식

````python
[ [0, 0, 1, 0, 0, 5], [1, 4, 0, 3, 2, 5], [0, 6, 0, 3, 0, 0], [2, 0, 0, 0, 0, 0],
 [0, 0, 0, 7, 0, 8], [1, 0, 0, 0, 0, 0]]
````

````python
from scipy import sparse

dense2 = np.array([[0, 0, 1, 0, 0, 5],
                   [1, 4, 0, 3, 2, 5],
                   [0, 6, 0, 3, 0, 0],
                   [0, 0, 0, 7, 0, 0],
                   [1, 0, 0, 0, 0, 0]])

# 0이 아닌 데이터 추출
data2 = np.array([1, 5, 1, 4, 3, 2, 5, 6, 3, 2, 7, 8,1])

# 행 위치와 열 위치를 각각 array로 생성
row_pos = np.array([0, 0, 1, 1, 1, 1, 1, 2, 2, 3, 4, 4, 5])
col_pos = np.array([2, 5, 0, 1, 3, 4, 5, 1, 3, 0, 3, 5, 0])

# COO 형식으로 변환
sparse_coo = sparse.coo_matrix((data2, (row_pos, col_pos)))

# 행 위치 배열의 고유한 값의 시작 위치 인덱스를 배열로 생성
row_pos_ind = np.array([0, 2, 7, 9, 10, 12, 13])

# CSR 형식으로 변환
sparse_csr = sparse.csr_matrix((data2, col_pos, row_pos_ind))

print('COO 변환된 데이터가 제대로 되었는지 다시 Dense로 출력 확인')
print(sparse_coo.toarray())
print('CSR 변환된 데이터가 제대로 되었는지 다시 Dense로 출력 확인')
print(sparse_csr.toarray())
````
