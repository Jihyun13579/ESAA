## 데이터 전처리

### 1. labelencoding

```python
# data 전처리 > 원본 data 분리하기

data = data_org.copy()
```

```python
data["sex"]
```

```python
# sex : female or male로 이루어진 categorical 변수이므로 labelencoder!
# female=0, male=1

from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
data['sex'] = le_sex.fit_transform(data['sex'])
data.head()
```

```python
data["smoker"]
```

```python
# smoker : yes or no로 이루어진 categorical 변수이므로 labelencoder!
# yes=1, no=0

from sklearn.preprocessing import LabelEncoder
le_smoker = LabelEncoder()
data['smoker'] = le_smoker.fit_transform(data['smoker'])
data.head()
```

```python
data["region"]
```

```python
data["region"].unique()   # 4가지 종류의 지역
```

```python
# region : 'southwest', 'southeast', 'northwest', 'northeast'로 이루어진 categorical 변수이므로 labelencoder!
# 'southwest'=3, 'southeast'=2, 'northwest'=1, 'northeast'=0

from sklearn.preprocessing import LabelEncoder
le_region = LabelEncoder()
data['region'] = le_region.fit_transform(data['region'])
data.head()
```
